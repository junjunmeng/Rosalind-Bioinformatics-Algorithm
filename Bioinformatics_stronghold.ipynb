```python
import pandas as pd
import numpy as np
```

```python
def Counting_DNA_Nucleotides(dna):
    """
    dna: the path to a file including DNA string
    output: the count of 'A', 'C', 'G', 'T'
    """
    
    with open(dna, 'r') as f:
        dna = f.readline()
    
    count_A = dna.count('A')
    count_C = dna.count('C')
    count_G = dna.count('G')
    count_T = dna.count('T')
    
    return count_A, count_C, count_G, count_T
    
```

```python
Counting_DNA_Nucleotides('./dataset/rosalind_dna.txt')
```

```python
def DNA_to_RNA(dna):
    """
    :dna: the path to dna file, which include dna string
    :return rna
    """
    
    with open(dna, 'r') as f:
        dna = f.read()
        
    rna = dna.replace('T', 'U')
    return rna
```

```python
DNA_to_RNA('./dataset/rosalind_rna.txt')
```
